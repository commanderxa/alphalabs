```python
import os

import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.ticker as ticker

# set nice theme for plots
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"] + plt.rcParams["font.serif"]
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["axes.ymargin"] = 0.0
plt.rcParams["axes.xmargin"] = 0.0
plt.rcParams['savefig.pad_inches'] = 0.01
```

```python
DPI = 600
```

```python
colors = ["red", "orange", "blue", "green", "purple"]
```

## 1D motion

```python
cm = 1 / 2.54
figsize = (8.5 * cm, 3.75 * cm)
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=figsize, dpi=DPI, sharex=True)
fig.subplots_adjust(wspace=1 * cm, hspace=0.2 * cm)

y_labels = [r"$\upsilon$ err, (log scale)", "x err, (log scale)"]
x_labels = ["Time, [s]", "Time, [s]"]
pattern = ["v", "x"]

dir_1d = "../errors/1d"

for i in range(2):
    j = 0
    for f in sorted(os.listdir(dir_1d)):
        if pattern[i] not in f or "_e_" in f:
            continue
        data = np.load(f"{dir_1d}/{f}")
        data1 = np.load(f"{dir_1d}/{f.replace('_rk_', '_e_')}")
        ts = float(f.split("_")[2])
        sns.lineplot(x=data[:, 1], y=data[:, 0], ax=ax[i], linewidth=1, label=None if i==0 else f"{ts:.0e} s", color=colors[j])
        sns.lineplot(x=data1[:, 1], y=data1[:, 0], ax=ax[i], linewidth=1, alpha=0.7, linestyle=":", color=f"tab:{colors[j]}")
        j += 1
    ax[i].set_ylabel(y_labels[i], fontsize=7, labelpad=2)
    ax[i].set_xlabel(x_labels[i], fontsize=7, labelpad=2)

    ax[i].tick_params(which="both", direction="in", labelsize=6, top=True, right=True, length=2)
    ax[i].yaxis.get_offset_text().set_fontsize(5)

    xticks = [0, 2, 4, 6, 8, 10]
    ax[0].set_xticks(xticks)
    ax[0].set_xticklabels([0, 2, 4, 6, 8, 10])

ax[0].set_yscale("log")
ax[1].set_yscale("log")
ax[0].set_ylim(9e-6, 0.12)

legend1 = ax[1].legend(fontsize=5, handlelength=1)
legend1.get_frame().set_linewidth(0.5)
ax[1].add_artist(legend1)

rk4_legend = mlines.Line2D([], [], color="black", linestyle="-", label="RK4")
euler_legend = mlines.Line2D([], [], color="gray", linestyle=":", label="Euler")
ax[1].legend(
    handles=[rk4_legend, euler_legend],
    loc="lower left",
    frameon=False,
    edgecolor="none",
    fontsize=5,
    handlelength=1,
    handleheight=0.15,
)

# ax[1].set_ylim(0, 0.01)
chart_name = f"1d_motion_rmse_analysis"
# if not IS_COLAB: 
chart_name = f"../output/" + chart_name
fig.savefig(f"{chart_name}.png", bbox_inches="tight")
```
